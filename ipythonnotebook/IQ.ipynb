```python
#IQ2 binary search
import math 
def binarysearch(arr,target):
    left=0
    right=len(arr)-1
    middle = int ((left+right)/2)
    
    while(left<right):
        if (target==arr[middle]):
            return middle
        if (target<arr[middle]):
            right=middle
            middle = int ((left+right)/2)
        else:
            left=middle
            middle = math.ceil((left +right)/2)
    return        
arr=[1,2,3,4,5,6,7,8,9]
binarysearch(arr,7)
            
            
        

```

```python
def binary_search(arr,left,right,key):
  if (left>right):
    return False
  middle=int((left + right)/2)  
  if(arr[middle]==key):
    return True
  if(arr[middle]<key):
    return binary_search(arr,middle+1,right,key)
  if(arr[middle]>key):
    return binary_search(arr,left,middle,key)
arr=[1,2,3,4,5,6,7,8,9]
binary_search(arr,0,8,1)    
```

```python
#IQ9 maximum differnce, Time complexity is O(n^2)
def MaximumDifference(lst):
    d_max=0
    for i in range(len(lst)):
        for j in range(i+1,len(lst)):
            if(lst[j]-lst[i]>d_max):
                d_max=lst[j]-lst[i]
    return d_max
lst=[5,8,9,21,3,1]

MaximumDifference(lst)
```

```python
#IQ18 sum closest to zero , o(n^2)
def SumClosestToZero(lst):
    minimum_sum=100
    for i in range(len(lst)):
        for j in range(i+1,len(lst)):
            if(abs(minimum_sum)>abs(lst[i]+lst[j])):
                minimum_sum=(lst[i]+lst[j])
    return minimum_sum
lst=[-2,9,6,1,2,-5]
SumClosestToZero(lst)               
              
    
```

```python
#IQ26 K contiguous array
def KcontiguousArray(lst,k):
    maxi=-1000
    for i in range(len(lst)-k+1):
        for j in range(i,i+k):
            if(maxi<lst[j]):
                maxi=lst[j]
        print(maxi)
        maxi=-1000
    return
lst=[8,6,1,12,48,56,15,24,39]
k=3
KcontiguousArray(lst,k)
```

```python
#strings IQ16  palindrome
def palindrome(phrase):
    reverse_phrase=reversed(phrase)
    if(list(phrase)==list(reverse_phrase)):
        print("given string is palindrome")
    else:
        print("is not palindrome")
    return
phrase="madam"
palindrome(phrase)
```
