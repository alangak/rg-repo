```python
#linklist IQ2 Reversing a list

class list_node:
    def __init__(self,value,Next=None):
        self.value=value
        self.Next=Next 

        
class link_list:
    def __init__(self,head=None):
        self.head=head
    def insert_no(self,value):   #inserting no at the end of list
        node=list_node(value)
        if(self.head==None):
            self.head=node
            return 
        else:
            curr_node=self.head
            while(curr_node.Next != None):
                curr_node=curr_node.Next
            curr_node.Next=node
    def reverse_list(self):
        pre_node=None
        curr_node=self.head
        next_node=self.head.Next
        while(curr_node):
            next_node=curr_node.Next
            curr_node.Next=pre_node
            pre_node=curr_node
            curr_node=next_node
        
        self.head=pre_node
        return
    def PrintList(self):
        curr_node=self.head
        while(curr_node):
            print(curr_node.value)
            curr_node=curr_node.Next
        return         








l3 = link_list()
l3.insert_no(35)
l3.insert_no(45)
l3.insert_no(50)
l3.insert_no(55)
l3.insert_no(60)
l3.PrintList()
l3.reverse_list()
l3.PrintList()







```
