```python
import numpy as np
import seaborn as sns
tips=sns.load_dataset('tips')
tips.head(5)
```

```python

tips.head(5)
```

```python
#distribution plot   #for single variable or univariate
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=40) #KDE is kernel density estimation
```

```python
#joint plot ....for bivariate
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
#it plots every possible combination colums
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex') #hue is actually adding categorical data
```

```python
sns.rugplot(tips['total_bill'])
```

```python
import numpy as np
#categorical plots
sns.barplot(x='sex',y='total_bill',data=tips)   #bydefault it is giving mean of total_bill by male amd female respectively
                                                 # i.e estimator is mean  
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)  #it gives no of occurences of male and female
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips) #datapoints are too much overlappig
                                                 # so we will use jitter
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)  #combination of violin and strip plot
                                                 #very less overlap of points,each points are easily visible
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)  #swarmplot on top of violin plot
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')   #general plot is factorplot
                                                               #rest of plots are special case of factor plots     
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
#MATRIX PLOTS
flights=sns.load_dataset('flights')
flights.head()
```

```python
#to use heat map(kind of matrix plot),we should have variables in index as well as column
tc=tips.corr()
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc,annot=True)
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
fp
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp,cmap='coolwarm')
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='black',linewidth='3')
```

```python
#cluster map          #we can see similar data are clustered together
sns.clustermap(fp,cmap='coolwarm')    #like april and october have similar kinda data
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)  #doing standardize the scale
```

```python
#GRIDS
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
g=sns.PairGrid(iris)
g
```

```python
import matplotlib.pyplot as plt
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
