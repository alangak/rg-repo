Quick Note on Super() and OOP

```python
class SimpleClass():
    
    def __init__(self,str_input):
        print("SIMPLE"+str_input)
```

```python
class ExtendedClass(SimpleClass):
    
    def __init__(self):
        print('EXTENDED')  
```

```python
s = ExtendedClass()
```

```python
class ExtendedClass(SimpleClass):
    
    def __init__(self):
        
        super().__init__(" My String")
        print('EXTENDED')
```

```python
s = ExtendedClass()
```
