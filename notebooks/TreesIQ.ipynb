```python
#IQ32 print vertical tree order ......time complexity-O(n^2)
class TreeNode:
    def __init__(self,value):  
        self.value=value
        self.left=None
        self.right=None
    def print_tree(self):
        if(self.left):
        
            self.left.print_tree()
            
        print(self.value)
        if(self.right):
            
            self.right.print_tree()
class TreeNode14(TreeNode):
    def find_min_max(self,curr_node,hd,lst):
        if(curr_node==None):
            return None
        if(hd<lst[0]):
            lst[0]=hd
    
        elif(hd>lst[-1]):
            lst.append(hd)
            
            
        self.find_min_max(curr_node.left,hd-1,lst) 
        self.find_min_max(curr_node.right,hd+1,lst)
        return lst
    def print_vertical_tree(self,curr_node,line,hd):
        if(curr_node==None):
            return None
        if(line==hd):
            print(curr_node.value)
        self.print_vertical_tree(curr_node.left,line,hd-1)    
        self.print_vertical_tree(curr_node.right,line,hd+1)
    def range_of_lines(self,curr_node,mini,maxi):
        for line in range(mini,maxi+1):
            self. print_vertical_tree(curr_node,line,0)
        return    
            
        
        
root=TreeNode14(1)
node2=TreeNode14(2)                                            #Tree 1-2,3   2-4,5    3-6,7   6-8,None
node3=TreeNode14(3)
node4=TreeNode14(4)
node5=TreeNode14(5)
node6=TreeNode14(6)
node7=TreeNode14(7)
node8=TreeNode14(8)
root.left=node2
root.right=node3
node2.left=node4
node2.right=node5
node3.left=node6
node3.right=node7
node6.left=node8  

lst=[0]
a=root.find_min_max(root,0,lst)
print(a)
mini=a[0]
maxi=a[-1]
root. range_of_lines(root,mini,maxi)
```

```python
#IQ38 check whether tree is sum tree or not
class TreeNode:
    def __init__(self,value):  
        self.value=value
        self.left=None
        self.right=None
    def print_tree(self):
        if(self.left):
        
            self.left.print_tree()
            
        print(self.value)
        if(self.right):
            
            self.right.print_tree()
class TreeNode16(TreeNode):
    def sum_of_tree(self,curr_node):
        if(curr_node==None):
            return False
        if(curr_node.left==None and curr_node.right==None):
            return curr_node.value
        l=self.sum_of_tree(curr_node.left)
        r=self.sum_of_tree(curr_node.right)
        if(curr_node.value==l+r):
            return 2*(l+r)
        return False
        
        
    
root=TreeNode16(50)
node2=TreeNode16(15)                                            #Tree 1-2,3   2-4,5    3-6,7   6-8,None
node3=TreeNode16(10)
node4=TreeNode16(10)
node5=TreeNode16(5)
node6=TreeNode16(7)
node7=TreeNode16(3)
root.left=node2
root.right=node3
node2.left=node4
node2.right=node5
node3.left=node6
node3.right=node7   


root.sum_of_tree(root)
    
```
