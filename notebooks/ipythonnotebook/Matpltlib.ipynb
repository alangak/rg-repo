```python

import matplotlib.pyplot as plt
plt.plot([2,4,6,4])
plt.ylabel("numbers")
plt.xlabel('indices')
plt.title('my plot')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.ylabel('squares')
plt.xlabel('numbers')
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'bo')
plt.ylabel('squares')
plt.xlabel('numbers')
plt.grid()
plt.show()
```

```python
#setp()....set property sets different property for different lines in same plot
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]
line=plt.plot(x1,y1,x2,y2)
plt.grid()
plt.setp(line[0],color='g',linewidth='1.0')
plt.setp(line[1],color='r',linewidth='3.0')
```
