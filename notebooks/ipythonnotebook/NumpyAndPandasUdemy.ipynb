```python
#MERGING and joining and concatenating
import pandas as pd
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
      'B':['B0','B1','B2','B3'],
      'C':['C0','C1','C2','C3'],
      'D':['D0','D1','D2','D3']
       },index=[0,1,2,3])
df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
      'B':['B4','B5','B6','B7'],
      'C':['C4','C5','C6','C7'],
      'D':['D4','D5','D6','D7']
       },index=[4,5,6,7])
df3=pd.DataFrame({'A':['A8','A9','A10','A11'],
      'B':['B8','B9','B10','B11'],
      'C':['C8','C9','C10','C11'],
      'D':['D8','D9','D10','D11']
       },index=[8,9,10,11])
print(df1)
print(df2)
print(df3)
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
left=pd.DataFrame({'A':['A0','A1','A2','A3'],
      'B':['B0','B1','B2','B3'],
      'key':['K0','K1','K2','K3']
       })
right=pd.DataFrame({

      'C':['C0','C1','C2','C3'],
      'D':['D0','D1','D2','D3'],
       'key':['K0','K1','K2','K3']

        }
        )
left
```

```python
pd.merge(left,right,how='inner',on='key')
```

```python
left=pd.DataFrame({'A':['A0','A1','A2','A3'],
      'B':['B0','B1','B2','B3'],
      'key1':['K0','K0','K1','K2'],
      'key2':['K0','K1','K0','K1']
       })
right=pd.DataFrame({

      'C':['C0','C1','C2','C3'],
      'D':['D0','D1','D2','D3'],
       'key1':['K0','K1','K1','K2'],
       'key2':['K0','K0','K0','K0']

        }
        )
pd.merge(left,right,how='inner',on=['key1','key2'])
```

```python
pd.merge(left,right,how='outer',on=['key1','key2'])
```

```python
pd.merge(left,right,how='right',on=['key1','key2'])
```

```python
#SF SALARIES
import pandas as pd
sal=pd.read_csv('Salaries.csv')
```

```python
import pandas as pd
sal['BasePay']=pd.to_numeric(sal.BasePay,errors='coerce')
sal['OvertimePay']=pd.to_numeric(sal.OvertimePay,errors='coerce')
sal['OtherPay']=pd.to_numeric(sal.OtherPay,errors='coerce')
sal['Benefits']=pd.to_numeric(sal.Benefits,errors='coerce')
```

```python
sal['BasePay'].mean()
```

```python
sal['OvertimePay'].max()
```

```python
sal[sal['EmployeeName']=="JOSEPH DRISCOLL"]['TotalPayBenefits']
```

```python
sal[sal['TotalPayBenefits']==sal['TotalPayBenefits'].max()]['EmployeeName'] # or sal['EmployeeName'][sal['TotalPayBenefits']==sal['TotalPayBenefits'].max()]
```

```python
#5 most common jobs
sum(sal[sal['Year']==2013]["JobTitle"].value_counts()==1)
```

```python
def chief_string(title):
    if 'chief'in title.lower().split():
        return True
    else:
        return False
sum(sal['JobTitle'].apply(chief_string))
    
```

```python
#CORRELATION BETWEEN JOB TITLE STRING AND TotalPayBenefits
sal['title_len']= sal["JobTitle"].apply(len)
sal[['TotalPayBenefits','title_len']].corr()
```
