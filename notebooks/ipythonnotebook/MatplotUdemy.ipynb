```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#Functional method
import numpy as np
x=np.linspace(0,5,11)
y=x**2
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.show()
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object orientated method
fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('y label')
axes.set_title('Title')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.4,0.15,0.4,0.3])
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("larger plot")


axes2.plot(y,x)
axes2.set_title("smaller plot")
```

```python
#SUB PLOT USING OBJECT ORIENTATED
fig,axes=plt.subplots()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
print(axes) #AXES is nothing but list of matplotlib axes So we can iterate through,in earlier case i.e
            #fig=plt.figure we had to define our own axes as fig.add_axes([.4,.2,.3,.4]) ,here we dont have to

```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)  
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)  
axes[0].plot(x,y)
axes[0].set_title('first plot')


axes[1].plot(y,x)
axes[1].set_title('second plot')
plt.tight_layout()  #this will make sure that plots are not overlapped
```

```python
#FIGURE SIZE AND DPI
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2),dpi=100) #Default dpi=100,we can changes it also

axes[0].plot(x,y)

axes[1].plot(y,x)
plt.tight_layout()
                      
```

```python
#to save a figure
fig.savefig("my_picture.png")
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3)
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label=' X squared')
ax.plot(x,x**3,label="X cube")

ax.legend()  #works only when labels are mentioned
```

```python
#PLOT APPEARANCE
fig = plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green',linewidth=3,linestyle= '--',marker='o',markersize=10,markerfacecolor='yellow',
         markeredgewidth=3,markeredgecolor='red' )
```

```python
fig = plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green',lw=3,ls= '--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
