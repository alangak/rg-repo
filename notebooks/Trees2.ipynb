```
#IQ49 form balanced binary tree from given sorted array...0(n)
import math
class TreeNode:
    def __init__(self,value):  
        self.value=value
        self.left=None
        self.right=None
    def print_tree(self):
        if(self.left):
        
            self.left.print_tree()
            
        print(self.value)
        if(self.right):
            
            self.right.print_tree()

def array_to_balanced_bst(lst,start,end):
  if(start>end):
    return None
  middle=math.floor((start+end)/2)
  curr_node=TreeNode(lst[middle])
  curr_node.left=array_to_balanced_bst(lst,start,middle-1)
  curr_node.right=array_to_balanced_bst(lst,middle+1,end)
  return curr_node

lst=[10,20,30,40,50,60,70]
root=array_to_balanced_bst(lst,0,6)
root.print_tree()




            
```

```
#Insertion of node in AVL tree
class TreeNode:
    def __init__(self,value):  
        self.value=value
        self.left=None
        self.right=None
        self.height=1
    def print_tree(self):
        if(self.left):
        
            self.left.print_tree()
            
        print(self.value)
        if(self.right):
            
            self.right.print_tree()
class TreeNode4(TreeNode):
  def insertion(self,curr_node,key):    #returns root of new AVL tree 

    if(curr_node==None):
      return TreeNode4(key)
    elif(key<curr_node.value):
      curr_node.left=self.insertion(curr_node.left,key)
    else:
      curr_node.right=self.insertion(curr_node.right,key)

    
    
    curr_node.height=1+max(self.getHeight(curr_node.left),self.getHeight(curr_node.right))

    
    bf=self.balance(curr_node)
    

    if(bf>1 and key<curr_node.left.value):      #Left Left
      return self.right_rotate(curr_node)
    if(bf>1 and key>curr_node.left.value):      #left right

      curr_node.left=self.left_rotate(curr_node.left)     
      return self.right_rotate(curr_node)
    if(bf<-1 and key>curr_node.right.value):      #right right
    
      return self.left_rotate(curr_node)
    if(bf<-1 and key<curr_node.right.value):      #right left 
      curr_node.right=self.right_rotate(curr_node.right)
      return self.left_rotate(curr_node)
      
    return curr_node

  def right_rotate(self,z):
    y=z.left
    T3=y.right
    #perform rotation
    y.right=z
    z.left=T3

    #update heights of ancestor nodes
    z.height=1+max(self.getHeight(z.left),self.getHeight(z.right)) 
    y.height=1+max(self.getHeight(y.left),self.getHeight(y.right))

    return y
  def left_rotate(self,z):
    y=z.right
    T2=y.left
    #perform rotation
    y.left=z
    z.right=T2

    #update heights of ancestor nodes
    z.height=1+max(self.getHeight(z.left),self.getHeight(z.right)) 
    y.height=1+max(self.getHeight(y.left),self.getHeight(y.right))
  

    return y

  def getHeight(self,curr_node):
    
    if(curr_node==None):
      return 0
    return curr_node.height

  def balance(self,curr_node):
    
    if(curr_node==None):
      return 0
    return (self.getHeight(curr_node.left)-self.getHeight(curr_node.right))

  def delete(self,curr_node,key):
    if (curr_node==None):
      return None
    elif(curr_node.value>key):
      curr_node.left= self.delete(curr_node.left,key)
    elif(curr_node.value<key):
      curr_node.right= self.delete(curr_node.right,key)
    else:
      if(curr_node.left==None and curr_node.right==None):
        return None
      elif(curr_node.left==None):
        temp=curr_node.right
        curr_node==None
        return temp  
      elif(curr_node.right==None):
        temp=curr_node.left
        curr_node==None
        return temp
      else:
        temp=self.getMinimumNode(curr_node.right)
        curr_node.value=temp.value
        curr_node.right=self.delete(curr_node.right,temp.value)
    
    curr_node.height=1+max(self.getHeight(curr_node.left),self.getHeight(curr_node.right))

    
    bf=self.balance(curr_node)
    

    if(bf>1 and key<curr_node.left.value):      #Left Left
      return self.right_rotate(curr_node)
    if(bf>1 and key>curr_node.left.value):      #left right

      curr_node.left=self.left_rotate(curr_node.left)     
      return self.right_rotate(curr_node)
    if(bf<-1 and key>curr_node.right.value):      #right right
    
      return self.left_rotate(curr_node)
    if(bf<-1 and key<curr_node.right.value):      #right left 
      curr_node.right=self.right_rotate(curr_node.right)
      return self.left_rotate(curr_node)
      
    return curr_node 
  def getMinimumNode(self,curr_node):
    while(curr_node.left != None):
      curr_node=curr_node.left
    return curr_node   






root=TreeNode4(1)
root=root.insertion(root,2)
root=root.insertion(root,3)
root=root.insertion(root,4)
root=root.insertion(root,5)
root=root.insertion(root,6)
root=root.insertion(root,7)
root=root.insertion(root,8)

root.delete(root,6)

root.print_tree()
```
