```
time complexity O(n^2)
def no_of_triangle(arr,n):
  arr.sort()
  print(arr)
  count=0
  for i in range(n-2):
    j=i+1
    k=i+2
    while(j<=n-2 and k<=n-1):
      if(arr[i]+arr[j]>arr[k]):
        count=count+1
        k=k+1
      else:
        if(j<k):
          j = j+1
        elif(j==k):
          k = k+1
  return count        
arr=[3,4,5,7,6,25,15]
n=7  
no_of_triangle(arr,n)            
```

```
def miss_no(arr,l,r,a,d):
  middle=int((l+r)/2)
  while(l<r):
    middle=int((l+r)/2)
    if(arr[middle]==a+middle*d):
      l=middle+1
    else:
      r=middle-1
  return a+middle*d
arr=[-10,10,20,30,40,50,60,70]
miss_no(arr,0,7,-10,10)        
    
```

```
#rotation of string
def if_rotated(str1,str2):
  n=len(str1)
  count=1
  lst=list(str1)
  while(count<=n):
    temp=lst[0]
    for i in range(1,n):
      lst[i-1]=lst[i]
    lst[n-1]=temp  
    print(''.join(lst))
    
    if(lst==list(str2)):
      print("they are rotated")
      break;
    count +=1  
  if(count==n):
    print("they are not rotated")   
if_rotated("ravi","irav")       
```

```
#min no of coins....bottom up approach
dp=[0]*100
def min_coins(coins,summ,n):
  dp[0]=0
  for i in range(1,summ+1):
    mini=10000
    for j in range(n-1,-1,-1):
      if(i-coins[j]>=0):
        dp[i]=1+dp[i-coins[j]]
        if(dp[i]<mini):
          mini=dp[i]
    dp[i]=mini      
        
  return dp[summ]
coins=[1,5,6,9]
n=len(coins)
min_coins(coins,11,n)        



```

```
#remove all adjacent duplicates
def remove_dup(s,n):
  stack=[s[0]]
  for i in range(1,n-1):
    if(len(stack)>0):

      if(s[i]==stack[-1] and s[i]==s[i+1]):
        stack.append(s[i])
      elif(s[i]==stack[-1] and s[i]!=s[i+1]):  
        while(len(stack)>0 and  s[i]==stack[-1]):
          stack.pop()
      else:
        stack.append(s[i])
    else:
      stack.append(s[i])    
  
  if(s[n-1]==stack[-1]):         #for last member of string
    while(len(stack)>0 and s[n-1]==stack[-1]):
      stack.pop()
  else:
    stack.append(s[n-1])    

  if(len(stack)>0):
    print("".join(stack))
  else:
    print("empty string") 
#s="azxxzy" 
#s="caaabbbaacdddd"     #result will be emoty string
#s="geeksforgeeg"
#s="acaaabbbacdddd"
s="mississipie"
n=len(s)
remove_dup(s,n)         

    

```

```
#element that appears once in sorted array....divide and conquer
def search(arr,l,r):
  if(l==r ):
    return -1
  mid=(l+r)//2
  if(arr[mid]!=arr[mid-1] and arr[mid]!= arr[mid+1]):
    return arr[mid]
  left=search(arr,l,mid)
  if(left !=-1):
    return left
  right=search(arr,mid+1,r)
  return right    
arr=[1,1,2,2,3,3,3,4,4,5,5,6,7]  
n=len(arr)
search(arr,0,n-1)
```
